# Pandas Basics using WHO Data

This notebook will go through the basic functionalities of Pandas using ** WHO ** Data

```python
%ls WHO.csv
```

```python
import pandas as pd
%matplotlib inline
```

## Read the CSV File

```python
WHO = pd.read_csv("WHO.csv", encoding="utf-8")
```

## View all the columns in a DataFrame

```python
list(WHO.columns)
```

## View the Summary / Description of the DataFrame

```python
WHO.describe()
```

### You can also view the summary / description of particular column within a DataFrame

```python
WHO['Over60'].describe()
```

## View the column data-types within the DataFrame

```python
WHO.dtypes
```

## View all the rows containing `NaN`s in any column

```python
WHO[WHO.isnull().any(axis=1)]

# Alternatively, you can do the same like so
WHO[pd.isnull(WHO).any(axis=1)]
```

## View all rows which have a null value under a specific column

```python
# Shows all records, where the LiteracyRate is Null
WHO[pd.isnull(WHO['LiteracyRate']) == True]

# Alternatively, you can also do the same like so
WHO[WHO['LiteracyRate'].isnull()]
```

### Contrastingly, you can select all the rows where a particular column is NOT Null

```python
WHO[~pd.isnull(WHO['LiteracyRate']) == True]
```

## View all rows, where there are NaNs in multiple columns

```python
WHO[(pd.isnull(WHO['LiteracyRate']) == True) & (pd.isnull(WHO['FertilityRate']) == True)]
```

### NOTE

To verify what nans look like, verify it against `np.nan` - like so...

```python
pd.isnull(pd.np.nan)
```

## View rows which only have `Europe` as a `Region`

** NOTE ** : The `==` is always mostly faster than the `.isin` method

```python
%timeit WHO[WHO["Region"] == "Europe"]

%timeit WHO[WHO["Region"].isin(["Europe"])]
```

## Compute `mean`, `sum`, `std` (standard deviation), etc. of a column

```python
print WHO['Under15'].mean()
print WHO['Under15'].std()
print WHO['Under15'].count()
```

## View row, for a column with a specific value

```python
# argmin returns the index of the row, with the min value for 'Under15' column
# then use the `iloc` to get the row of that index
print WHO.iloc[WHO['Under15'].argmin()]

# and if you directly want to view the actual `min` value, then do the following
min_value = WHO['Under15'].min()
print "\n ===== \n Min Value: {}".format(min_value)
```

## View row by index

** NOTE ** - Almost, NEVER use the `ix` method, if you want to get a row by it's index. Only use the `iloc` - which takes only an integer value i.e. the index, where as the `loc` method takes any value only if that column that you are searching against is an indexed column.

Example Below

```python
# Search directly by index (viewing 85th row)
print WHO.iloc[85]

# `loc` also takes index (viewing 85th row)
print WHO.loc[85]

# as well as any value if the column is index
WHO_indexed = WHO.set_index('Under15')
print WHO_indexed.loc[13.12]
```

### Similarily, viewing the row with the max value against a particular column

```python
WHO.iloc[WHO['Under15'].argmax()]
```
